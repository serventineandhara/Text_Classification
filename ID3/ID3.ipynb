# Data Mining

## Decision Tree (ID3)

### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

### Preparing the data

```python
df_path = 'Data/News_dataset.csv'
df = pd.read_csv(df_path,sep=';')
df.head()
```

```python
train = df.copy()
```

### Preprocessing

*using label encoder*

```python
encoder = LabelEncoder()
```

```python
train.shape
```

```python
length = train.shape[1]
col = train.columns
for i in range (length):
    print(i)
    a = encoder.fit_transform(train.iloc[:,i:i+1])
    a = pd.DataFrame(a, columns=[col[i]+'new'])
    train= train.join(a)
```

```python
train
```

```python
test = train
```

### ID3 Algorithm

```python
clf = DecisionTreeClassifier(
    criterion= "entropy"
)

clf_entropy = clf.fit(train.iloc[:,3:5], train.iloc[:,5])
```

```python
clf_entropy
```

```python
pred = clf_entropy.predict(test.iloc[:,3:5])
print("predicted values = ")
print(pred)
```

```python
predict = []
for i in pred:
    if i == 0:
        predict.append("bussiness")
    elif i == 1:
        predict.append("entertainment")
    elif i == 2:
        predict.append("politics")
    elif i == 3:
        predict.append("sport")
    else :
        predict.append("tech")
```

```python
predict
```

```python
result = test.iloc[:,0:3]
```

```python
result['predict'] = predict
```

```python
result
```

### Evaluation

#### Confusion Matrix

```python
conf_matrix = confusion_matrix(test.iloc[:,5], pred)
print("Confusion Matrix: ", conf_matrix) 
      
#     print ("Accuracy : ", 
#     accuracy_score(y_test,y_pred)*100) 
      
#     print("Report : ", 
#     classification_report(y_test, y_pred)) 
```

#### F-measure

```python
precision = conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[1][0])
recall = conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[0][1])
fmeasure = (2*precision*recall)/(precision+recall)
```

```python
print("Precision : ", precision)
print("recall : ", recall)
print("f-measure : ", fmeasure)
```

#### Accuracy

```python
acc = accuracy_score(test.iloc[:,5], pred)*100
print(acc,"%")
```

#### Report

```python
print("report :")
print(classification_report(test.iloc[:,5], pred) )
```
